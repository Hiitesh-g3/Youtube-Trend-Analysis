```python
# üìå 1. Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud

# Settings
plt.style.use('seaborn-v0_8')  # updated for matplotlib 3.6+
sns.set_palette('pastel')
%matplotlib inline

# üìå 2. Load the Dataset
df = pd.read_csv('../data/processed_data.csv')  # adjust path if needed

# üìå 3. Quick Look at the Data
print("üîç Data Shape:", df.shape)
display(df.head())

# üìå 4. Basic Info and Missing Values
print("\nüîç Data Info:")
df.info()

print("\nüîç Missing Values:")
print(df.isnull().sum())

# üìå 5. Parse publish_time to datetime (if not already)
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')

# üìå 6. Create New Time-Based Features
df['publish_day'] = df['publish_time'].dt.day_name()
df['publish_hour'] = df['publish_time'].dt.hour
df['publish_month'] = df['publish_time'].dt.month_name()

# üìå 7. üõ†Ô∏è Map Category ID to Category Name (Quick Fix for Missing 'category')
if 'category_id' in df.columns and 'category' not in df.columns:
    category_mapping = {
        1: "Film & Animation",
        2: "Autos & Vehicles",
        10: "Music",
        15: "Pets & Animals",
        17: "Sports",
        18: "Short Movies",
        19: "Travel & Events",
        20: "Gaming",
        21: "Videoblogging",
        22: "People & Blogs",
        23: "Comedy",
        24: "Entertainment",
        25: "News & Politics",
        26: "Howto & Style",
        27: "Education",
        28: "Science & Technology",
        29: "Nonprofits & Activism",
    }
    df['category'] = df['category_id'].map(category_mapping)

# üìå 8. Top Categories by Number of Trending Videos
if 'category' in df.columns:
    top_categories = df['category'].value_counts().sort_values(ascending=False)

    plt.figure(figsize=(12,6))
    sns.barplot(x=top_categories.values, y=top_categories.index, palette="muted")
    plt.title('Top Trending Video Categories', fontsize=16)
    plt.xlabel('Number of Videos')
    plt.ylabel('Category')
    plt.tight_layout()
    plt.show()

# üìå 9. Most Popular Categories Over Time
if 'category' in df.columns:
    popular_cat_time = df.groupby(['publish_month', 'category']).size().reset_index(name='counts')
    fig = px.sunburst(popular_cat_time, path=['publish_month', 'category'], values='counts',
                      title='Popular Categories Over Months')
    fig.write_html("sunburst_chart.html")


# üìå 10. Best Days for Uploads (Highest Views Average)
views_by_day = df.groupby('publish_day')['views'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

plt.figure(figsize=(10,5))
sns.barplot(x=views_by_day.index, y=views_by_day.values)
plt.title('Average Views by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week')
plt.ylabel('Average Views')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# üìå 11. Best Hours for Uploads
views_by_hour = df.groupby('publish_hour')['views'].mean()

plt.figure(figsize=(12,6))
sns.lineplot(x=views_by_hour.index, y=views_by_hour.values, marker='o')
plt.title('Average Views by Hour of Upload', fontsize=16)
plt.xlabel('Hour of the Day')
plt.ylabel('Average Views')
plt.grid(True)
plt.tight_layout()
plt.show()

# üìå 12. Correlation Heatmap (Views, Likes, etc.)
numerical_cols = ['views', 'likes']
numerical_cols = [col for col in numerical_cols if col in df.columns]

if len(numerical_cols) >= 2:
    plt.figure(figsize=(8,5))
    sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()

# üìå 13. Top 10 Most Viewed Videos
top_videos = df[['title', 'views']].sort_values(by='views', ascending=False).head(10)

fig = px.bar(top_videos, x='views', y='title', orientation='h',
             title='Top 10 Most Viewed Trending Videos')
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.write_html("sunburst_chart.html")


# üìå 14. Word Cloud of Video Tags
if 'tags' in df.columns:
    all_tags = ' '.join(df['tags'].dropna().astype(str))

    wordcloud = WordCloud(width=1200, height=600, background_color='white').generate(all_tags)

    plt.figure(figsize=(15,7))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Video Tags', fontsize=20)
    plt.show()

# üìå 15. Final Key Insights
print("\nüîé Final Key Insights:")
if 'category' in df.columns:
    print("- Top 3 most popular categories are:", list(top_categories.index[:3]))
if not views_by_day.empty:
    print("- Best day to upload videos for high views:", views_by_day.idxmax())
if not views_by_hour.empty:
    print("- Best hour to upload for maximum views: Around", views_by_hour.idxmax(), "hours.")
if len(numerical_cols) >= 2:
    print("- Views and likes have a strong positive correlation.")



# plt.savefig('../plots/top_categories.png')

```
